```python
# import necessary modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# TheMovieDB.org

```python
# Creating DataFrame from The Movie DB budget file
tn_budgets_df = pd.read_csv("data/tn.movie_budgets.csv.gz")
tn_budgets_df.head()
```

```python
tn_budgets_df.info()
```

# IMDB Title Basics

```python
# Creating DataFrame from IMDB Title Basics file
imdb_title_basics_df = pd.read_csv('data/imdb.title.basics.csv.gz')
imdb_title_basics_df.head()
```

```python
imdb_title_basics_df.info()
```

```python
# Renaming columns to prepare for a later merge with tn_budgets
imdb_title_basics_df = imdb_title_basics_df.rename(columns = {'primary_title': 'movie', "release_date": "start_year"})
imdb_title_basics_df.head()
```

```python
# Adding a "start_year" column to prepare for merge with imdb_title_basics
tn_budgets_df["start_year"] = tn_budgets_df.release_date.map(lambda x: int(x[-4:]))
tn_budgets_df.head()
```

```python
# Merging on <movie> and <start_year>
# Note: Merging on <start_year> fixed some issues with duplicates and missing data.  
total_df = pd.merge(imdb_title_basics_df, tn_budgets_df, on=["movie", "start_year"])
```

```python
# Dropping "original_title" column since it is all duplicates and unnecessary
total_df = total_df.drop("original_title", axis=1)
```

```python
# Removing dollar signs and commas from monetary values
total_df.production_budget = total_df.production_budget.map(lambda x: x.replace("$", ""))
total_df.production_budget = total_df.production_budget.map(lambda x: x.replace(",", ""))

total_df.domestic_gross = total_df.domestic_gross.map(lambda x: x.replace("$", ""))
total_df.domestic_gross = total_df.domestic_gross.map(lambda x: x.replace(",", ""))

total_df.worldwide_gross = total_df.worldwide_gross.map(lambda x: x.replace("$", ""))
total_df.worldwide_gross = total_df.worldwide_gross.map(lambda x: x.replace(",", ""))
```

```python
# Casting monetary values to int from str
total_df.production_budget = total_df.production_budget.map(lambda x: int(x))
total_df.domestic_gross = total_df.domestic_gross.map(lambda x: int(x))
total_df.worldwide_gross = total_df.worldwide_gross.map(lambda x: int(x))
```

```python
# Adding column for Profit
total_df['Profit'] = total_df.worldwide_gross - total_df.production_budget
```

```python
# Removed outliers
# NOTE: It was our decision to not include any movie that made less than $100,000
# worldwide since it was more likely that such a movie would either be (1) a very 
# large negative outlier and/or (2) a data entry/missing data problem
total_df = total_df.loc[total_df.worldwide_gross > 100000]
```

```python
total_df.sort_values("Profit", ascending=False)
```

```python
# Splitting genres column into individual genre columns so we can access one genre
# at a time
genres_df = total_df.genres.str.split(",", 2, expand=True)

total_df["genre1"] = genres_df[0]
total_df.head()
total_df["genre2"] = genres_df[1]
total_df["genre3"] = genres_df[2]
```

```python
# Dropping unnecessary columns
total_df = total_df.drop("genres", axis=1)
total_df = total_df.drop("tconst", axis=1)
total_df = total_df.drop("id", axis=1)
```

```python
# Creates three-letter abbreviations for release month
total_df['release_month'] = total_df.release_date.map(lambda x: x[:3])
```

```python
# This function returns a numerical representation of a three-letter
# month abbreviation
def month_to_num(x):
   if x == "Jan":
       x = 1
   elif x == "Feb":
       x = 2
   elif x == "Mar":
       x = 3
   elif x == "Apr":
       x = 4
   elif x == "May":
       x = 5
   elif x == "Jun":
       x = 6
   elif x == "Jul":
       x = 7
   elif x == "Aug":
       x = 8
   elif x == "Sep":
       x = 9
   elif x == "Oct":
       x = 10
   elif x == "Nov":
       x = 11
   elif x == "Dec":
       x = 12
   else:
       x = None
   return x
```

```python
# Using function to change release month to an integer
total_df.release_month = total_df.release_month.map(lambda x: month_to_num(x))
```

```python
# Remove movie duplicates
total_df = total_df[~total_df.duplicated("movie")]
```

```python
# Using groupby to get mean profit by month
# Note: The x/1000000 changes values to dollar amounts in millions - this improves readability

release_month_means = total_df.groupby("release_month").mean()["Profit"].map(lambda x: x/1000000)
mean_roi_month = plt.bar(release_month_means.index, release_month_means, color = 'skyblue')
plt.xlabel("Month")
plt.ylabel("Mean Profit (in Millions USD)")
plt.title("Mean Profit by Month")
plt.xticks(range(1,13), ("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"));
```

# Mean Profit by Month Conclusions

Months that return the highest profits are May-July, followed by November and December.  This intuitively makes sense.

May through July are the months for the big summer blockbusters.  Kids are out of school and it's a great time to push movies with big budgets and big star power.  November and December are not only the holiday season but also the last bit of time to push for Oscar contention.

Furthermore, January (and to a smaller extent September and October) are "dump months".  Dump months are months when movie studios typically push films they are contractually obligated to release but less confident in.  There will be less high-level competition during this time.

Altogether, big movies with big stars and big names attached to them may do well competing in those summer months.  However, movies without the name recognition or the big budgets may actually be better in months like January, September or October when the competition is weaker.

```python
# This function determines mean profit of all movies in DataFrame <df> with genre <genre>
def genre_mean_ROI(df, genre_list):
    row_means = []
    for genre in genre_list:
        genre_filter = df.loc[(df["genre1"] == genre) | (df["genre2"] == genre) | (df["genre3"] == genre)]
        row_means.append(genre_filter.mean()["Profit"])
    return row_means
```

```python
# List of unique genres from our DataFrame
genre_list = ['Action', 'Adventure', 'Comedy', 'Horror', 'Drama', 'Biography', 'Crime', 'Animation', 'Fantasy', 'Documentary', 'Mystery', 'Thriller', 'Musical', 'Romance', 'Sci-Fi', 'Family', 'Music', 'Western', 'History', 'Sport', 'War']
```

```python
# Create smaller DataFrame with Genre and Mean Profit

d = list(zip(genre_list, genre_mean_ROI(total_df, genre_list)))
roi_genre_df = pd.DataFrame(d, columns=['Genre','Mean ROI'])
```

```python
# This function determines mean domestic gross of all movies in DataFrame df
# with genre <genre>
def genre_mean_DG(df, genre_list):
    row_means = []
    for genre in genre_list:
        genre_filter = df.loc[(df["genre1"] == genre) | (df["genre2"] == genre) | (df["genre3"] == genre)]
        row_means.append(genre_filter.mean()["domestic_gross"])
    return row_means
```

```python
# Creating international gross column in DataFrame
total_df['intl_gross'] = total_df.worldwide_gross - total_df.domestic_gross
```

```python
# This function determines mean international gross of all movies in DataFrame df
# with genre <genre>
def genre_mean_IG(df, genre_list):
    row_means = []
    for genre in genre_list:
        genre_filter = df.loc[(df["genre1"] == genre) | (df["genre2"] == genre) | (df["genre3"] == genre)]
        row_means.append(genre_filter.mean()["intl_gross"])
    return row_means
```

```python
# Create smaller DataFrame with Genre and Mean Profit

total_df.head()
d = list(zip(genre_list, genre_mean_ROI(total_df, genre_list)))
roi_genre_df = pd.DataFrame(d, columns=['Genre','Mean Profit'])

# Create small domestic gross DataFrame
dg = list(zip(genre_list, genre_mean_DG(total_df, genre_list)))
dg_genre_df = pd.DataFrame(dg, columns=['Genre','Mean DG'])
sorted_dg_genre_df = dg_genre_df.sort_values(by='Mean DG', ascending=False)

# Create small international gross DataFrame
ig = list(zip(genre_list, genre_mean_IG(total_df, genre_list)))
ig_genre_df = pd.DataFrame(ig, columns=['Genre','Mean IG'])
sorted_ig_genre_df = ig_genre_df.sort_values(by='Mean IG', ascending=False)

# Plot Mean Profit by Genre

sorted_roi_genre_df = roi_genre_df.sort_values(by='Mean Profit', ascending=False)
plt.bar(sorted_roi_genre_df.Genre, sorted_roi_genre_df['Mean Profit']/1000000, color='skyblue')
plt.xticks(rotation = "vertical");
plt.ylabel('Mean Profit (in millions USD)')
plt.xlabel('Movie Genre')
plt.title('Mean Profit by Genre');
```

# Mean Profit by Genre Conclusions

The animation, sci-fi, adventure and musical genres all provide a high mean profit, with animation reigning supreme.  We'll look at some violin plots below to examine this a bit further.

```python
# Plot Mean Domestic Gross by Genre
sorted_dg_genre_df = dg_genre_df.sort_values(by='Mean DG', ascending=False)
plt.bar(sorted_dg_genre_df.Genre, sorted_dg_genre_df['Mean DG']/1000000, color = 'skyblue')
plt.xticks(rotation = "vertical");
plt.ylabel('Mean Domestic Gross (in millions USD)')
plt.xlabel('Movie Genre')
plt.title('Domestic Gross by Genre')
plt.ylim(0.0, 263.1403457041237);
```

# Domestic Gross by Genre Conclusions

Again, the top four genres are the same.  Mostly this graph was created to compare to International Gross numbers, so this graph never made it into the final deck.

```python
# Create DataFrame to plot international gross
d = list(zip(genre_list, genre_mean_IG(total_df, genre_list)))
ig_genre_df = pd.DataFrame(d, columns=['Genre','Mean IG'])

# Plot international gross vs genre
sorted_ig_genre_df = ig_genre_df.sort_values(by='Mean IG', ascending=False)
plt.bar(sorted_ig_genre_df.Genre, (sorted_ig_genre_df['Mean IG']/1000000), color = 'skyblue')
plt.xticks(rotation = "vertical");
plt.ylabel('International Gross (in millions USD)')
plt.xlabel('Movie Genre')
plt.title('International Gross by Genre')
plt.ylim()
```

# International Gross by Genre Conclusions

After comparing to domestic gross, there doesn't seem to be a whole lot of surprising data here.  Domestic and international gross both show pretty much the same order with some very minor changes.  

This graph also didn't make it into the final deck because we decided that having a stacked bar plot would be (1) more visually interesting and (2) still capable of showing that the same conclusion - that there is not enough of a differences between domestic and foreign appetites to consider it.

```python
# Create small domestic gross DataFrame
dg = list(zip(genre_list, genre_mean_DG(total_df, genre_list)))
dg_genre_df = pd.DataFrame(dg, columns=['Genre','Mean DG'])
sorted_dg_genre_df = dg_genre_df.sort_values(by='Mean DG', ascending=False)

# Create small international gross DataFrame
ig = list(zip(genre_list, genre_mean_IG(total_df, genre_list)))
ig_genre_df = pd.DataFrame(ig, columns=['Genre','Mean IG'])
sorted_ig_genre_df = ig_genre_df.sort_values(by='Mean IG', ascending=False)
```

```python
# Merge above DataFrames and create stacked bar plot
dg_ig_df = pd.merge(sorted_dg_genre_df, sorted_ig_genre_df, on=['Genre'])

dg_ig_df['Mean DG'] = dg_ig_df['Mean DG']/1000000
dg_ig_df['Mean IG'] = dg_ig_df['Mean IG']/1000000

sns.set()
sns.set_style("ticks")
stacked_plot = dg_ig_df.set_index('Genre').plot(kind='bar', stacked=True, color=['skyblue','dodgerblue'])
stacked_plot.set(ylim=(0, 400))
stacked_plot.set(xlabel='Movie Genre', ylabel='Domestic/Int\'l Gross (in Millions USD)')
stacked_plot.set_title('Domestic/Int\'l Gross by Genre')
stacked_plot.legend(['Domestic', 'International']);
```

# Domestic/Int'l Gross by Genre

This graph was the one that made it into the deck.  It shows small an dinteresting deviations, such as the Fantasy genre being more popular outside of domestic audiences or Documentaries making much money across the board, but the main takeaway is thet domestic and international audiences seem to appreciate the same genres.

```python
# Creating violin plot for genre Adventure
adv_violin = total_df.loc[(total_df.genre1 == 'Adventure') | (total_df.genre2 == 'Adventure') | (total_df.genre3 == 'Adventure')]

adv_plot = sns.violinplot(y=adv_violin.Profit/1000000)

adv_plot.set_title('Profit for Adventure Genre (in millions USD)');
ymin, ymax = adv_plot.get_ylim() #Info on y limits so all violinplots can have the same y axis
adv_plot.set_ylim(bottom=-732.7127319149596, top=2052.8457727667305);
```

# Profit for Adventure Genre Conclusions

The full bar here shows a large range for the "non-outlier" movies.  Should Microsoft's offering fall in this range, there is a high probability it will make good money.  

Also, the positive outliers are gigantic.  If Microsoft's movie exceeds expectations, it is possible to have stratospheric returns.

Of the top 4 genres, this is one that we recommend.
